```python
api = "http://fund.eastmoney.com/js/fundcode_search.js"
res = requests.get(api)
```

```python
funds = json.loads(res.content.decode("utf-8")[9:-1])
funds = [{"id": id, "code": code, "name": name, "kind": kind, "fullcode": fullcode} for id, code, name, kind, fullcode in funds]
```

```python
funds
```

```python
def has_dividends(fund_id):
    res = requests.get(f"https://fundf10.eastmoney.com/F10DataApi.aspx?type=lsjz&code={fund_id}&page=1&per=50&sdate=2020-09-10&edate=2021-09-10")
    return "分红送配" in res.content.decode("utf-8")
```

```python
fund_id = "000001"
res = requests.get(f"https://fundf10.eastmoney.com/F10DataApi.aspx?type=lsjz&code={fund_id}&page=1&per=50&sdate=2020-09-10&edate=2021-09-10")
```

```python
funds_with_dividends = [fund for fund in funds[:10] if has_dividends(fund["id"])]
```
